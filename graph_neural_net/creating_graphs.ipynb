# Create graphs with NetworkX

```python
import networkx as nx

graph = nx.DiGraph()
graph.add_nodes_from([
    (1, {'color': 'red', 'size': 10}),
    (2, {'color': 'blue', 'size': 20}),
    (3, {'color': 'green', 'size': 30}),
    (4, {'color': 'yellow', 'size': 40}),
    (5, {'color': 'purple', 'size': 50})
])

graph.add_edges_from([
    (1, 2),
    (2, 3),
    (3, 4),
    (4, 5),
    (5, 1),
    (2, 4)
    ])
node_colors = nx.get_node_attributes(graph, 'color').values()
node_sizes = nx.get_node_attributes(graph, 'size').values()
nx.draw(graph, with_labels=True, node_color=list(node_colors), node_size=list(node_sizes))


node_sizes
```

```python
und_graph = graph.to_undirected()
nx.draw(und_graph, with_labels=True, node_color=list(node_colors), node_size=list(node_sizes))
```
